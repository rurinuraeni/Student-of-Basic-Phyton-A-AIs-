# Soal 1

###### Buatlah sebuah program yang dapat menyimpan kontak dan juga menampilkan kontak-kontak  tersebut. Setiap kontak terdiri dari nama dan nomor telepon. Saat menjalankan program, program akan menampilkan pesan sambutan dan juga menampilkan menu. Program akan secara otomatis kembali menu apabila user telah selesai melihat kontak ataupun menambahkan kontak.

```python
def menu():
    print("Selamat datang!")
    print("---Menu---")
    print("1. Daftar Kontak")
    print("2. Tambah Kontak")
    print("3. Keluar")

nama3={}
nomor3={}

while True:
        print (' ')
        menu()
        print (' ')
        
        a = input("Pilih menu: ")
        num=3
        if a=="2":
            nama3=[]
            nomor3=[]
            for i in range (num):
                nama=input("Nama: ")
                nomor=int(input("No. Telepon: "))
                nama3.append(nama)
                nomor3.append(nomor)
                print("Kontak berhasil disimpan")
                print(' ')
        elif a=="1":
            print("Daftar Kontak")
            for i in range (len(nama3)):
                print("Nama: " + str(nama3[i]))
                print("No. Telepon: "+ str(nomor3[i]))
                print(' ')  
        elif a=="3":
            print("Program selesai, sampai jumpa!")
            break
        else:
            print("Menu tidak tersedia")   
    
```

##### Buatlah program interaktif yang akan menerima n-angka dari user,mengecek apakah angka yang diterima valid atau tidak, dan menghitung dari rata-rata angka yang telah dimasukkan.  Jika input dari user tidak valid, sistem akan tetap meminta inputan ke user untuk angka ke-n hingga data yang diterima valid.

```python
try:
    num=int(input('Masukan jumlah angka yang dimasukkan: '))
    angka2 = 0
    if num < 1 or num>1:
        for i in range (num):
            y=float(input('Masukan angka: '))
            angka2 += y
        rata_rata1 = angka2/num
        print ('Rata-rata angka yang dimasukkan adalah: '+ str((round(rata_rata1))))
except:
    print('Data yang anda masukkan tidak valid. Silahkan masukan angka kembali!')




```
